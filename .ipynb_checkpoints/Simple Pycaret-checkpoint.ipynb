# IMPORTS 

```python
import numpy as np
import pandas as pd
import pycaret
import numpy as np
import matplotlib.pyplot as plt
from pycaret.classification import * 
from sklearn.model_selection import train_test_split
```

# LOAD DATA

```python
# READ IN TRAIN
entire = pd.read_csv('./data/fundraising.csv')

# SPLIT INTO TRAIN AND DEV
train, dev = train_test_split(entire, test_size=0.20, random_state=42)

# READ IN TEST
test = pd.read_csv('./data/future_fundraising.csv')

# MAKE TARGET NANs
test['target'] = np.nan
```

```python

# Assuming 'df' is your DataFrame containing the data
df = pd.read_csv('./data/fundraising.csv')  # Replace with your actual data loading method if needed

# Interaction between 'income' and 'wealth'
df['interaction_wealth_income'] = df['income'] * df['wealth']

# Income to Home Value Ratio
df['income_to_home_ratio'] = df['income'] / df['home_value']

# Wealth to Home Value Ratio
df['wealth_to_home_ratio'] = df['wealth'] / df['home_value']

# Gift to Income Ratio
df['gift_to_income_ratio'] = df['lifetime_gifts'] / df['income']  # Assuming 'lifetime_gifts' is the total amount donated

# Average Gift Size to Largest Gift Ratio
df['average_gift_size_to_largest_gift_ratio'] = df['avg_gift'] / df['largest_gift']

# Total Gifts to Number of Promotions
df['total_gifts_to_number_of_promotions'] = df['lifetime_gifts'] / df['num_prom']

# You can now use 'df' with these new features to train your model or further analysis
```

```python
# =============================================================================
# PYCARET 
# =============================================================================
from pycaret.classification import ClassificationExperiment, setup, compare_models, tune_model

exp = ClassificationExperiment()
exp.setup(train, target = 'target', session_id = 27)

best = compare_models()
```

```python
from pycaret.classification import setup, compare_models, create_model, tune_model, evaluate_model, finalize_model, predict_model, plot_model

# Correctly setting up the experiment with proper parameters
exp_setup = setup(data=train, target='target', session_id=27, fold_strategy='stratifiedkfold', fold=10, verbose=True)

# Creating an AdaBoost model
ada = create_model('ada')

# Tuning AdaBoost parameters
tuned_ada = tune_model(ada, optimize='AUC', n_iter=50)

# Evaluating the tuned model
evaluate_model(tuned_ada)

# Finalizing the model
final_ada = finalize_model(tuned_ada)
```
